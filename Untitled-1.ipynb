### Parte I: Estructuras de Datos
**Implemente ejemplos propios donde:**\
Cree y manipule listas (append, insert, remove, sort). Use slicing.\
Cree diccionarios y recorra sus claves y valores.\
Cree sets y demuestre operaciones de unión e intersección.\

```python
# Creacion y manipulacion de listas
ciudades = ['Santo Domingo', 'Minocqua', 'La Vega', 'Bonao']
print(f'Lista original: {ciudades}\n')

ciudades.append('Bonao')
ciudades.insert(1, 'Santiago')
ciudades.remove('La Vega')
ciudades.sort()

print(f'Lista modificada: {ciudades}')
print(f'Slicing de la lista: {ciudades[0:2]}')

# Se creo una lista, se le agregaron elementos con append e insert, se borro uno con remove, se ordeno con sort y se sacaron los primeros dos con slicing.
```

```python
# Diccionarios
trabajos = {
    'mañana': 'call center',
    'tarde': 'laboratorio',
    'noche': 'universidad'
}

for clave, valor in trabajos.items():
    print(f'En la {clave} estoy en el {valor}')

# Se creo un diccionario con mis horarios y se recorrio con un for para imprimir las claves y valores.
```

```python
# Sets
grupo1 = {'juan', 'francis', 'paola'}
grupo2 = {'francis', 'luis', 'ana'}

print(f'Union: {grupo1 | grupo2}')
print(f'Interseccion: {grupo1 & grupo2}\n')

# Comprehensions
numeros = [1, 2, 3, 4]
cuadrados = [x**2 for x in numeros]
print(f'List comprehension: {cuadrados}')

nombres = ['ledwin', 'francis']
largo_nombres = {n: len(n) for n in nombres}
print(f'Dict comprehension: {largo_nombres}')

# Se hicieron operaciones con sets, 
# y se usaron comprehensions para crear una lista y un diccionario 
# de forma rapida en una sola linea.
```

### Parte II: Funciones
**Implemente:**\
Una función con múltiples parámetros.\
Una función con parámetro por defecto.\
Una función que retorne múltiples valores.\
Una función lambda.\
Uso de map(), filter() y sorted() con función personalizada.\
Explique el propósito de cada función.

```python
# funcion multiples parametros
def multiplicar(a, b):
    return a * b

print(f'Resultado multiplicar: {multiplicar(5, 4)}')
# esta funcion toma dos numeros y los multiplica.\

# funcion parametro por defecto
def saludo(nombre, saludo="Hola"):
    print(f'{saludo}, {nombre}')

saludo('Ledwin')
# esta funcion saluda, si no le pasas el saludo, dice Hola por defecto.

# funcion multiples valores
def operaciones(a, b):
    return a+b, a-b

suma, resta = operaciones(10, 5)
print(f'Suma: {suma}, Resta: {resta}')
# esta funcion devuelve la suma y la resta de dos numeros al mismo tiempo.

# funcion lambda, map, filter, sorted
doble = lambda x: x * 2

numeros = [5, 2, 8, 1, 9]
filtrados = list(filter(lambda x: x > 3, numeros))
mapeados = list(map(doble, filtrados))
ordenados = sorted(mapeados)

print(f'Procesados con lambda, filter, map y sorted: {ordenados}')
# se uso lambda para multiplicar por 2,
# filter para sacar los mayores de 3, map para aplicar el lambda, 
# y sorted para ordenar de menor a mayor.
```

### Parte III: Manejo de Archivos
**Cree un archivo.txt.**\
Escriba contenido en él.\
Lea el contenido.\
Procese las líneas del archivo.\
Use with open() correctamente.\
Explique la importancia del manejo adecuado de archivos.

```python
# Manejo de archivos
with open('archivo_ledwin.txt', 'w') as f:
    f.write('Hola, este es mi archivo.\n')
    f.write('Estoy haciendo la practica 2.\n')

with open('archivo_ledwin.txt', 'r') as f:
    lineas = f.readlines()
    for linea in lineas:
        print(linea.strip())

# Es importante manejar bien los archivos con with open() 
# para que se cierren solos automaticamente al terminar de leer o escribir, 
# asi no gastan memoria ni se corrompen los datos.
```
