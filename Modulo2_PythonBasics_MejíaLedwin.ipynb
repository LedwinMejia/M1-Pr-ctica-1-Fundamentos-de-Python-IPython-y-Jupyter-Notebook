### 2.2 Estructuras de Datos
Cree:\
**Una lista\
Una tupla\

Un set**\
Acceda a elementos por índice.\
Modifique valores.\

```python
# Creacion a acceso a elementos de listas, tuplas, diccionarios y sets:

# lista
edades = [34, 56, 11, 89]
print(f'Lista: {edades[3]}')

# tupla
frutas = ('pera', 'manzana', 'mango', 3)
print(f'Tupla: {frutas[2]}')

# sets
personas = {'marcos', 45, 'kathy', 34, 'juana', 42}
# no es posible acceder a los elementos de un set mediante indice 

# diccionario
profecion = {
    'ledwin': 'programador',
    'michael': 'cantante',
    'richard': 'asistente',
    'carlos': 'camarero'
}
print(f'Diccionario: {profecion['ledwin']}')

# DataFrame
tabla_sets = pd.DataFrame(personas, columns=['Datos'])
```

```python
# Modificar valores

# Lista
edades[3] = 55

# diccionario
profecion['ledwin'] = ['cocinero']

print(edades)
print(profecion)
```

```python
# Recorra una lista usando un for.

for i in edades:
    if i < 18:
        print(f'Hay un menor de edad, y su edad es {i}.')
    else:
        pass
```

#### 2.3 Control de Flujo
***Escriba un bloque if-elif-else.\
Cree un ciclo for.\
Cree un ciclo while.\
Use comprensión de listas***.

```python
# if-elif-else
if len(edades) > 3:
    print(f'Este es un grupo de {len(edades)}, por lo tanto puede entrar.')
else:
    print(f'Este grupo es de 3 o menos personas, pueden entrar.')

# ciclo for
for i in profecion.values():
    print(i)

# while
x = 10
total = 0

while x > 0:
    if total > 40:
        break
    total += x
   
print(total)
    
# comprension de listas
numeros = [3, 6, 10]

numeros_cuadrado = [i ** 2 for i in numeros]
print(numeros_cuadrado)
```

### 2.4 Funciones

***Cree una función simple que:***\
Reciba dos parámetros.\
Retorne un resultado.\
Cree una función con valor por defecto.\
Documente la función con un docstring.

```python
# primera funcion
def nombre_apellido(nombre, apellido):
    """
    nombre de la funcion:
        nombre_apellido

    Esta funcion toma el nombre y apellido del usuario como parametros e imprime una cadena 
    de texto dicendo lo siguiente:  'Mi nombre es {nombre} y mi apellido es {apellido}'

    ejemplo:
        nombre_apellido('Ledwin', 'Mejia')

        resultado: Mi nombre es Ledwin y mi apellido es Mejia

    """
    print(f' Mi nombre es {nombre} y mi apellido es {apellido}')

print(nombre_apellido.__doc__)

nombre_apellido('Ledwin', 'Mejia')
nombre_apellido('Carlos', 'Jackson')


# segunda funcion
def suma_100(numero, cien=100):
    print(f'El resultado es {numero + cien}.')
    

suma_100(100)
```
